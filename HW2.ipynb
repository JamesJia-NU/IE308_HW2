#### Associations 

```python
frequentItemsets = apriori(finalData, min_support=0.001, use_colnames=True)
```

```python
frequentItemsets.shape
```

```python
print(frequentItemsets)
```

```python
rules = association_rules(frequentItemsets, metric="lift", min_threshold=1)
```

```python
print(rules)
```

```python
rules.describe()
```

#### RESULT EXPLORATION 

```python
## Sort rules by lift
rules.drop(["antecedent support","consequent support","leverage","conviction"], axis=1, inplace=True)
rules = rules.sort_values(by=['lift'], ascending=False)
rules.head(20)
```

```python
## LOOPER: put the items in the rules from rules (sorted by lift)
rank_list = []
looper = 1

for index, row in rules.iterrows():
    
    looper = looper + 1
    
    a = list(row.antecedents)
    c = list(row.consequents)
    for item in a:
        rank_list.append(item)
    for item in c:
        rank_list.append
```

```python
## remove duplicates in looper
rank_list_clean = list(dict.fromkeys(rank_list))
```

```python
len(rank_list_clean)
```

```python
finalResult = rank_list_clean[0:100]
len(finalResult)
```

```python
## Results exploration:
## convert the results to value
finalResultNum = []

for sku in finalResult:
    a = sku[4:]
    a_num = int(a)
    finalResultNum.append(a_num)

finalResultNum
    
```
